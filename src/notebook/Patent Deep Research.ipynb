** Disinfection with Plasma: A Technical Report Based on Patent Literature

*** Abstract (Summary)

This report synthesizes information from a collection of patent documents pertaining to disinfection using plasma technology. The inventions described address various challenges in disinfection, including the need for rapid, safe, cost-effective, and environmentally friendly methods applicable to a wide range of materials and environments. The patents cover diverse applications, from sterilizing medical instruments and disinfecting personal items to purifying water and air, highlighting the versatility of plasma-based disinfection techniques.

*** Background and Existing Technologies

Traditional disinfection methods, such as chemical disinfectants, autoclaving, and UV radiation, have limitations. Chemical disinfectants can leave harmful residues, contribute to antibiotic resistance (US20220265879A1), and may be ineffective against certain pathogens. Autoclaving, while effective, can damage heat-sensitive materials (US20220409758A1). UV radiation may not reach all surfaces and can be harmful to humans (WO2023041962A1). Plasma technology offers an alternative by generating reactive species that can effectively inactivate microorganisms without the drawbacks of conventional methods. Several patents address these limitations by providing innovative plasma-based disinfection solutions.

*** Technical Fields of Invention

The patents reviewed encompass a variety of technological areas related to disinfection with plasma:

*   **Plasma Generation:** Devices and methods for generating plasma, including Dielectric Barrier Discharge (DBD) (WO2021229615A1, US20230346998A1), microwave plasma (US20220409758A1, US9308285B2), radio frequency (RF) plasma (US12083237B2), and hydrated plasma (WO2023130525A1).
*   **Disinfection Systems:** Integrated systems for disinfecting specific items or environments, such as medical devices (US20150246149A1, EP4196179B1), herbs (US12083237B2, US20240324617A1), hands (WO2022063446A1, EP4221769A1), and water (US10604422B2, US11053140B2).
*   **Plasma-Activated Media:** Methods for creating plasma-activated water or other liquids for disinfection purposes (WO2021223986A3, US20220265879A1, US20240350633A1).
*   **Air Purification:** Devices for air disinfection and purification using plasma technology (WO2023041962A1, US10881757B2).
*   **Surface Disinfection:** Systems and methods for disinfecting surfaces using plasma (US20230346998A1, US11825841B2).

*** Inventions Related to Disinfection with Plasma

**** Rapid Disinfection Device (WO2021229615A1)

*   **Novelty and Objectives:** This invention introduces a rapid disinfection device utilizing Pulsed Electric Field (PEF) and Dielectric Barrier Discharge (DBD) technologies. The objective is to provide a fast, safe, and cost-effective alternative to traditional disinfection methods.
*   **Unique Components:** The device features a compartment for items to be disinfected, conductive plates (at least one electrically insulated), a high-voltage generator, and a control system.
*   **Technical Problem Solved:** Addresses the slow and potentially harmful nature of existing disinfection methods.
*   **Improvement:** Reduces disinfection time to seconds, is operable by untrained users, and is fully automatic.

**** Stethoscope Disinfection System (US20150246149A1)

*   **Novelty and Objectives:** Aims to provide a safe and efficient method for disinfecting stethoscopes using plasma energy to reduce nosocomial infections.
*   **Unique Components:** Includes an enclosure, electrodes, and a voltage supply to generate plasma within the enclosure.
*   **Technical Problem Solved:** Current stethoscope designs are difficult to sterilize without damage.
*   **Improvement:** Enables disinfection and sterilization of stethoscopes without risking structural damage.

**** Herb Disinfection Device (US12083237B2, US20240324617A1)

*   **Novelty and Objectives:** Provides a device and method for disinfecting herbs in small to medium quantities using plasma technology, suitable for home or retail use.
*   **Unique Components:** Includes a sealable vessel for containing herbs, a power source for generating plasma, and electrodes to apply an electromagnetic field for plasma generation.
*   **Technical Problem Solved:** Microbial contamination in herbs, especially for home use where industrial sterilization is absent.
*   **Improvement:** Reduces or eliminates harmful microorganisms in herbs while preserving their characteristics.

**** Hand Disinfection Device (WO2022063446A1, EP4221769A1)

*   **Novelty and Objectives:** An improved device for disinfecting hands using a plasma source in an air stream, with effective generation and transport of reactive plasma species.
*   **Unique Components:** A fan for generating airflow and a plasma generator with a dielectric tube, an electrically conductive core, and a wire wound around the outside to form electrodes.
*   **Technical Problem Solved:** The need for more efficient plasma source utilization for hand drying and disinfection.
*   **Improvement:** Provides simultaneous drying and disinfection with efficient use of plasma sources.

**** Microwave-Plasma Disinfector (US20220409758A1)

*   **Novelty and Objectives:** A versatile disinfector that combines microwave and plasma technology to decontaminate heat-sensitive materials more effectively and at lower temperatures.
*   **Unique Components:** A microwave source with multi-feed points and variable output power, and a low-temperature atmospheric-pressure plasma source.
*   **Technical Problem Solved:** Limitations of existing disinfection methods, including chemical shortages, material damage, and high energy consumption.
*   **Improvement:** Achieves more effective sterilization at lower temperatures, reduces material damage, and provides uniform energy distribution.

**** Plasma-Activated Water Disinfection Device (WO2021223986A3)

*   **Novelty and Objectives:** A device for disinfecting parts, especially body parts like hands, that can operate with minimal interruptions by producing plasma-activated water.
*   **Unique Components:** A manufacturing unit that produces plasma-activated water, which is then supplied to the pressurizing means as a disinfectant.
*   **Technical Problem Solved:** Existing hand disinfection devices require frequent refills and pose disinfection risks over extended use.
*   **Improvement:** Allows for independent operation with fewer interruptions and reduces the risk of disinfection issues.

**** Air and Surface Disinfection Device (WO2023041962A1)

*   **Novelty and Objectives:** A device that disinfects air and surfaces without producing harmful chemicals like ozone, hydrogen ions, or OH radicals.
*   **Unique Components:** A device case, a device fan with a timer function, and a device plasma consisting of 4 separate sections using nano atmospheric oxygen plasma.
*   **Technical Problem Solved:** Providing clean air and surface disinfection without the harmful side effects of existing technologies like UV radiation and ozone gas.
*   **Improvement:** A disinfection method that is safe for humans and effective in eliminating pollutants.

**** Antimicrobial Wipes (US11825841B2)

*   **Novelty and Objectives:** Methods and systems for generating antimicrobial wipes effective against bacteria, viruses, spores, and fungi for cleaning and disinfecting surfaces.
*   **Unique Components:** A non-thermal plasma generator and a feed system to move wipes past the plasma generator.
*   **Technical Problem Solved:** Surface contamination by bacteria, viruses, spores, and fungi.
*   **Improvement:** Activates the liquid within the wipe, creating antimicrobial properties.

**** Water Disinfection System (US10604422B2, US11053140B2)

*   **Novelty and Objectives:** A portable water container with an integrated, non-chemical disinfection system that eliminates bacteria, viruses, organic compounds, and emergent contaminants using plasma.
*   **Unique Components:** A source of high voltage power, electrodes, a high voltage power source controller, and an activation device.
*   **Technical Problem Solved:** Emergent contaminants in water that traditional disinfection methods may not effectively eliminate.
*   **Improvement:** Effective disinfection of water through plasma-induced reactions.

*** Applicability and Uses

The inventions described have broad applicability across various sectors:

*   **Healthcare:** Sterilization of medical instruments (US20150246149A1, EP4196179B1), disinfection of stethoscopes (US20150246149A1), and surface disinfection in hospitals (US20230346998A1).

*   **Air Purification:** Air disinfection in homes, offices, and public transportation (WO2023041962A1, US10881757B2).
*   **Agriculture:** Treatment of fresh produce to control postharvest diseases (WO2019040991A1).
*   **Medical Treatments:** Treatment of viral and bacterial infections of the respiratory tract (US20240350633A1).

*** Conclusion

The patent documents reviewed demonstrate a wide range of innovative approaches to disinfection using plasma technology. These inventions address the limitations of traditional disinfection methods by offering rapid, safe, cost-effective, and environmentally friendly solutions. The applications span diverse fields, highlighting the versatility and potential of plasma-based disinfection in various industries and settings. ** Plasma Technology for Tooth and Teeth Whitening: A Technical Report

*** Abstract (Summary)

This report analyzes patent literature pertaining to the application of plasma technology for tooth and teeth whitening. The inventions detailed herein focus on addressing limitations of traditional whitening methods, such as the use of high concentrations of hydrogen peroxide, and improving the efficacy and safety of the whitening process. The core innovations involve the use of plasma devices to generate reactive species in situ, the creation of cold atmospheric plasma (CAP)-containing solutions, and the development of portable, hand-held plasma devices for targeted treatment. These technologies aim to enhance tooth whitening, sterilization, and overall oral health.

*** Background and Existing Technologies

Traditional tooth whitening methods often rely on the application of high concentrations of hydrogen peroxide or other bleaching agents. While effective, these methods can cause tooth sensitivity, gum irritation, and other adverse effects. Furthermore, existing plasma generation systems can be bulky, require high power, or generate undesirable byproducts like ozone. The inventions discussed in this report seek to overcome these limitations by utilizing plasma technology to achieve safer, more efficient, and more targeted tooth whitening and oral care.

*** Technical Fields of Invention

The inventions discussed in this report span several technological areas:

*   **Plasma Generation:** Devices and methods for generating plasma, including non-thermal plasma, cold atmospheric plasma (CAP), and microwave-induced plasma (WO2000060910A1, US20120107761A1, US11771770B2).
*   **Dental Devices:** Integration of plasma technology into dental instruments and apparatus for clinical use (EP2308415A4, US20230056196A1).
*   **Surface Treatment:** Application of plasma for surface processing, material removal, and deposition on teeth (EP2552340A4, WO2015083155A1).
*   **Reactive Species Generation:** Methods for generating reactive oxygen and nitrogen species (RONS) using plasma for targeted treatment (EP2552340A4, WO2015083155A1).

*** Inventions Related to Plasma Technology for Tooth and Teeth Whitening

**1. Plasma-Based Tooth Whitening in Liquid Environment (EP2552340A4):**

*   **Novelty and Objectives:** This invention introduces a method for whitening teeth by generating plasma effluent in a liquid environment. The primary objective is to reduce the risks associated with high concentrations of hydrogen peroxide used in conventional methods.
*   **Unique Components/Methods:** The method involves submerging a portion of the tooth in a liquid, positioning a plasma device nearby, supplying ionizable media to the device, and igniting the media to form a plasma effluent. This effluent reacts with the liquid to produce reactive species.
*   **Technical Problem Solved:** Addresses the safety concerns of traditional whitening methods.
*   **Improvement:** Reduces safety risks by generating reactive species in situ that selectively interact with foreign matter on the tooth, leading to whitening.

**2. Cold Atmospheric Plasma (CAP)-Containing Solutions (US11771770B2):**

*   **Novelty and Objectives:** This invention focuses on creating CAP-containing solutions and systems for their production. The objective is to provide a versatile treatment modality for various biomedical applications, including teeth bleaching/whitening, by dissolving CAP in a solvent.
*   **Unique Components/Methods:** The system comprises a gas source, a plasma generating device with a hollow body fluidically coupled to the gas source, and a container for housing a fluid. The plasma generating device discharges CAP directly into the fluid.
*   **Technical Problem Solved:** Overcomes the limitations of direct CAP application, which can be limited in scope and invasiveness.
*   **Improvement:** Enables broader and less invasive applications of CAP, including teeth whitening, through the use of CAP-containing solutions.

**3. Non-Thermal Plasma Applicator for Oral Treatment (WO2015083155A1):**

*   **Novelty and Objectives:** This invention provides an apparatus and method for generating and applying non-thermal plasma for oral treatments, including teeth whitening.
*   **Unique Components/Methods:** The plasma applicator is configured to generate non-thermal plasma in a region adjoining a surface portion of a living body. It includes RF electrodes, a gas supply conduit, and a suction conduit.
*   **Technical Problem Solved:** Existing methods generate plasma at a distance from the treatment zone, reducing the effectiveness of short-lived reactive species.
*   **Improvement:** Generates plasma in close proximity to the treatment zone, enhancing the efficiency of teeth whitening and other oral treatments. The suction conduit removes fluids and limits the spread of active gaseous species.

**4. Hand-Held Non-Thermal Gas Plasma Device (US20120107761A1):**

*   **Novelty and Objectives:** This invention aims to provide a portable, hand-held device for generating non-thermal gaseous species, suitable for applications like cleaning and whitening teeth.
*   **Unique Components/Methods:** The device comprises a gas capsule, a reaction generator, a source of electrical energy, energizing means, and a housing, all configured for hand-held operation.
*   **Technical Problem Solved:** Existing gas plasma generators are either large industrial systems or smaller systems with bulky gas cylinders and power cables.
*   **Improvement:** Integrates all components into a single, hand-held housing, resulting in a portable, self-contained device for targeted treatment.

**5. Dental Disease Treating Device with Ozone Minimization (US20230056196A1):**

*   **Novelty and Objectives:** This invention provides a dental disease treating device that generates high-density plasma while minimizing ozone generation, suitable for treating dental diseases like periodontitis.
*   **Unique Components/Methods:** The device comprises a handpiece body, a plasma generating part, and a detachable tip that irradiates plasma locally while inhibiting oxygen from entering the plasma generating space.
*   **Technical Problem Solved:** The need for a device that generates high-density plasma for dental treatments while avoiding ozone generation.
*   **Improvement:** Generates plasma suitable for dental treatments with reduced ozone production, enhancing safety.

**6. Dental Clinical Apparatus with Plasma Jet Application (EP2308415A4):**

*   **Novelty and Objectives:** This invention provides a dental clinical apparatus that integrates a plasma jet applying means with a dental treatment instrument.
*   **Unique Components/Methods:** The apparatus includes a syringe with a fluid jetting system and a plasma jet applying means. The plasma jet is produced by applying a low-frequency high voltage between electrodes on a rare gas tube.
*   **Technical Problem Solved:** Conventional dental instruments lack the function of plasma jet application for improved wettability and sterilization.
*   **Improvement:** Improves wettability and sterilizes affected areas during treatment through plasma jet application.

**7. Method for Plasma Generation Using Microwave Electromagnetic Fields (WO2000060910A1):**

*   **Novelty and Objectives:** This invention ensures stable and reliable discharge generation to form a continuous plasma flow in any gas environment under any gas pressure using low-power microwave sources with automatic plasma initiation.
*   **Unique Components/Methods:** The method involves initiating a discharge in a gas medium by means of electromagnetic microwave, utilizing a low-power microwave source and a cavity resonator.
*   **Technical Problem Solved:** Prior art devices require high-intensity electric fields, leading to costly microwave devices with short lifespans.
*   **Improvement:** Ensures stable and reliable discharge generation, enabling continuous plasma flow using low-power microwave sources.

*** Applicability and Uses

The inventions described have a wide range of practical applications in dentistry:

*   **Tooth Whitening:** All inventions can be used for teeth whitening by removing stains and foreign matter from the tooth enamel (EP2552340A4, US11771770B2, WO2015083155A1, US20120107761A1).
*   **Sterilization:** Plasma jets and CAP solutions can sterilize infected tissues and inactivate microorganisms in the oral cavity (US11771770B2, EP2308415A4).
*   **Surface Treatment:** Plasma can be used to treat tooth surfaces prior to bonding, improving adhesion (WO2015083155A1).
*   **Periodontitis Treatment:** The dental disease treating device (US20230056196A1) is specifically designed for treating periodontitis.
*   **Wound Healing:** CAP solutions can promote wound healing in the oral cavity (US11771770B2).
*   **Hydrophobic Coatings:** Plasma treatment can create hydrophobic coatings on teeth to prevent bacterial growth (EP2552340A4).

The patents analyzed demonstrate a significant interest in utilizing plasma technology for tooth and teeth whitening and overall oral health.
